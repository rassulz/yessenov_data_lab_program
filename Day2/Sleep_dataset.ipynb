# Sleep health and lifestyle dataset analyze

```python
import pandas as pd

df = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")
print(df.shape)
df.head(200)
```

### Which gender sleep longer?

```python
avg_sleep = df.groupby("Gender")["Sleep Duration"].mean().round(2)
print(avg_sleep)

winner = avg_sleep.idxmax()
print(f"\n{winner}s sleep longer on average: {avg_sleep.max()} hours")
```

**Females sleep longer on average: 7.23 hours**

### Which gender has better sleep quality?

```python
# Which gender has better sleep quality?
avg_quality = df.groupby("Gender")["Quality of Sleep"].mean().round(2)
print(avg_quality)

best = avg_quality.idxmax()
print(f"\n{best}s have better sleep quality on average: {avg_quality.max()} / 10")
```

**Females have better sleep quality on average: 7.66 / 10**

### Which occupation sleeps longer and shorter?

```python
# Which occupation sleeps longer and shorter?
by_job = df.groupby("Occupation")["Sleep Duration"].mean().round(2).sort_values(ascending=False)
print(by_job)

print(f"\nLongest sleepers:  {by_job.idxmax()} ({by_job.max()} hours)")
print(f"Shortest sleepers: {by_job.idxmin()} ({by_job.min()} hours)")
```

### Which age group sleeps longer and shorter?

```python
# Which age group sleeps longer and shorter?
df["Age Group"] = pd.cut(df["Age"], bins=[20, 30, 40, 50, 60],
                         labels=["20s", "30s", "40s", "50s"], right=False)

by_age = df.groupby("Age Group", observed=True)["Sleep Duration"].mean().round(2).sort_values(ascending=False)
print(by_age)

print(f"\nLongest sleepers:  people in their {by_age.idxmax()} ({by_age.max()} hours)")
print(f"Shortest sleepers: people in their {by_age.idxmin()} ({by_age.min()} hours)")
```

### Correlation table between all the health / lifestyle numbers

```python
# Correlation table between all the health / lifestyle numbers
num_cols = ["Age", "Sleep Duration", "Quality of Sleep",
            "Physical Activity Level", "Stress Level", "Heart Rate", "Daily Steps"]

corr = df[num_cols].corr().round(2)
corr  # display as a table (closer to +1 = move together, closer to -1 = opposite)
```

### Occupation is a category, so we compare AVERAGES of each health metric per major

```python
# Occupation is a category, so we compare AVERAGES of each health metric per major
by_occupation = df.groupby("Occupation")[num_cols].mean().round(1)
by_occupation = by_occupation.sort_values("Quality of Sleep", ascending=False)
by_occupation
```

### BMI category

```python
# All BMI categories and how many people are in each
bmi_counts = df["BMI Category"].value_counts()
print(bmi_counts)
```

```python
# Which major tends to have which BMI category?
# (merge the duplicate "Normal Weight" label into "Normal" first)
df["BMI Category"] = df["BMI Category"].replace({"Normal Weight": "Normal"})

# Row % = of each occupation, what share falls in each BMI category
bmi_by_job = (pd.crosstab(df["Occupation"], df["BMI Category"], normalize="index") * 100).round(0)
print(bmi_by_job)

# The BMI category each major tends to have most
print("\nMost common BMI per occupation:")
print(bmi_by_job.idxmax(axis=1))
```

### Which gender tends to have insomnia?

```python
# Which gender tends to have insomnia?
# Count of insomnia cases per gender
insomnia = df[df["Sleep Disorder"] == "Insomnia"]
insomnia_count = insomnia["Gender"].value_counts()
print("Number of insomnia cases per gender:")
print(insomnia_count)

# Share = of each gender, what % suffer from insomnia
insomnia_rate = (df["Sleep Disorder"].eq("Insomnia")
                 .groupby(df["Gender"]).mean() * 100).round(1)
print("\n% of each gender that has insomnia:")
print(insomnia_rate)

winner = insomnia_rate.idxmax()
print(f"\n{winner}s tend to have insomnia more: {insomnia_rate.max()}% of them, "
      f"vs {insomnia_rate.min()}% of {insomnia_rate.idxmin()}s")
```

**Males tend to have insomnia more: 21.7% of them have it (41 cases), vs 19.5% of Females (36 cases).**

### Correlation between blood pressure and heart rate

```python
# Correlation between blood pressure and heart rate
# Blood Pressure is text like "126/83", so split it into two numbers first
df[["Systolic", "Diastolic"]] = df["Blood Pressure"].str.split("/", expand=True).astype(int)

bp_hr_corr = df[["Systolic", "Diastolic", "Heart Rate"]].corr().round(2)
print(bp_hr_corr)

print(f"\nSystolic  vs Heart Rate: {bp_hr_corr.loc['Systolic', 'Heart Rate']}")
print(f"Diastolic vs Heart Rate: {bp_hr_corr.loc['Diastolic', 'Heart Rate']}")
print("(closer to +1 = rise together, closer to -1 = opposite, near 0 = no link)")
```

**Blood pressure and heart rate have only a weak positive correlation: Systolic vs Heart Rate = 0.29, Diastolic vs Heart Rate = 0.27.** They tend to rise together slightly, but the link is weak. (Systolic and Diastolic, on the other hand, are almost perfectly correlated at 0.97.)

### Correlation between stress and sleep quality

```python
# Correlation between stress and sleep quality
stress_sleep_corr = df["Stress Level"].corr(df["Quality of Sleep"]).round(2)
print(f"Stress Level vs Quality of Sleep: {stress_sleep_corr}")
print("(closer to -1 = strong opposite link: more stress -> worse sleep)")
```

**Stress and sleep quality have a strong negative correlation: -0.90.** This is the strongest relationship in the whole dataset — the more stressed a person is, the worse their sleep quality. Stress is the biggest enemy of good sleep here.

## 📊 Summary — What this data tells us about sleep

After exploring **374 people** and their sleep, health, and lifestyle, a clear story comes out:

### 😟 Stress is the #1 enemy of sleep
The strongest relationship in the entire dataset is between **stress and sleep quality: −0.90**. That's almost a perfect opposite link — the more stressed a person is, the worse they sleep. Stress also drags down sleep *duration* (−0.81) and raises *heart rate* (+0.67).

### 😴 Good sleep comes as a package
People who sleep **longer** also sleep **better** (+0.88). Sleep length and sleep quality go hand in hand — you rarely get one without the other.

### 👩 vs 👨 Gender differences
- **Women** sleep a bit longer (7.23h vs 7.04h) and rate their sleep higher (7.66 vs 6.97).
- But **men** are slightly more likely to suffer from **insomnia** (21.7% vs 19.5%).

### 💼 Your job affects your sleep
- **Engineers** sleep best of all (8.4 / 10), are the least stressed, and stay mostly at a normal weight.
- **Sales Representatives** sleep the worst (4 / 10), are the most stressed, take the fewest steps (3,000/day), and 100% fall into the obese category.

### ❤️ A surprise: blood pressure ≠ heart rate
We might expect blood pressure and heart rate to rise together, but the link is **weak (0.29)**. Knowing one tells you little about the other in this group.

### 🏃 Movement matters
More physically active people also walk more (steps vs activity = +0.77) — the two reinforce each other.

> *Lower your stress and you'll sleep longer **and** better — and habits like your job and daily movement quietly shape your sleep just as much.*
